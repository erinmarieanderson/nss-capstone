### Graphs to Accompany "A Proposal to Update Boston Marathon Qualifying Times Using Age-Grading" 6/26/20

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("white") #don't want gridlines
%matplotlib inline
```

```python
plt.rc('font', size=22)          
plt.rc('axes', labelsize=28)    
plt.rc('figure', titlesize=44)
plt.rcParams['font.sans-serif'] = "Lucida Sans Unicode"
```

#### Part I. Single-Age Records and Proposed 2020 Age Standards

```python
record_times=pd.read_excel('../Too_Large/marathon_record_times.xlsx',
                 sheet_name='age records vs 2020 standards ')
```

```python
#Women's Single-Age Marathon Records
plt.figure(figsize=(20,10))
ax = sns.scatterplot(x="Age", y="Age_record_min", 
                     data=record_times[record_times['After_2015'].isin(['older_than_15'])], s=120, color ="darkorange", linewidth=.6, edgecolor="black")
ax = sns.scatterplot(x="Age", y="Age_record_min",
                     data=record_times[record_times['After_2015'].isin(['since_15'])], s=160, color ="darkorange", marker= "^", linewidth=.6, edgecolor="black")
ax = sns.scatterplot(x="Age", y="Age_record_min",
                     data=record_times[record_times['Age'].isin([25])], s=800, color ="yellow", marker='*', linewidth=2, edgecolor="darkorange")

plt.legend(labels=['Single-Age Records', 'Single-Age Records (> 1/1/15)', 'World Record'], frameon=False, bbox_to_anchor=(.4, .8), borderaxespad=0.)
plt.xlabel('Age')
plt.ylabel('Time (min)')
sns.despine(offset=10, trim=False)

plt.savefig('1__womens_single_age_marathon_records.png');
```

```python
#Women's Single-Age Marathon Records + 2020 Proposed Age Standards Curve
plt.figure(figsize=(20,10))
ax = sns.scatterplot(x="Age", y="Age_record_min", 
                     data=record_times[record_times['After_2015'].isin(['older_than_15'])], s=120, color ="darkorange", linewidth=.6, edgecolor="black")
ax = sns.scatterplot(x="Age", y="Age_record_min",
                     data=record_times[record_times['After_2015'].isin(['since_15'])], s=160, color ="darkorange", marker= "^", linewidth=.6, edgecolor="black")
ax = sns.lineplot(x="Age", y="2020_AS_min",
                     data=record_times, color ="darkorange", linewidth=5)
ax = sns.scatterplot(x="Age", y="Age_record_min",
                     data=record_times[record_times['Age'].isin([25])], s=800, color ="yellow", marker='*', linewidth=2, edgecolor="darkorange")

plt.legend(labels=['Proposed Age Standards', 'Single-Age Records', 'Single-Age Records (> 1/1/15)', 'World Record'], frameon=False, bbox_to_anchor=(.4, .8), borderaxespad=0.)
plt.xlabel('Age')
plt.ylabel('Time (min)')
sns.despine(offset=10, trim=False)

plt.savefig('2__womens_single_age_marathon_records_AND_2020_proposed_AS_curve.png');
```

```python
#Women's Single-Age Marathon Records + 2020 Proposed Age Standards Curve + 2015 Existing Age Standards Curve
plt.figure(figsize=(20,10))
ax = sns.scatterplot(x="Age", y="Age_record_min", 
                     data=record_times[record_times['After_2015'].isin(['older_than_15'])], s=120, color ="darkorange", linewidth=.6, edgecolor="black")
ax = sns.scatterplot(x="Age", y="Age_record_min",
                     data=record_times[record_times['After_2015'].isin(['since_15'])], s=160, color ="darkorange", marker= "^", linewidth=.6, edgecolor="black")
ax = sns.lineplot(x="Age", y="2020_AS_min",
                     data=record_times, color ="darkorange", linewidth=3)
ax = sns.lineplot(x="Age", y="2015_AS_min",
                     data=record_times, color ="red", linewidth=3)
ax = sns.scatterplot(x="Age", y="Age_record_min",
                     data=record_times[record_times['Age'].isin([25])], s=800, color ="yellow", marker='*', linewidth=2, edgecolor="darkorange")

plt.legend(labels=['Proposed Age Standards', '2015 Age Standards', 'Single-Age Records', 'Single-Age Records (> 1/1/15)', 'World Record'], frameon=False, bbox_to_anchor=(.4, .8), borderaxespad=0.)
plt.xlabel('Age')
plt.ylabel('Time (min)')
sns.despine(offset=10, trim=False)

#plt.savefig('womens_single_age_marathon_records_AND_2020_proposed_AS_curve_AND_existing_2015_AS_curve.png');
```

#### Part II. 2019 Boston Finishing Times

```python
bos_19=pd.read_excel('../Too_Large/bos_data.xlsx',
                 sheet_name='bos_19', skiprows=1)
```

```python
#Actual Finish Time by Age (Men & Women, Boston 2019)
sns.lmplot(x="Age", y="Official_Time_minutes", hue="Gender", data=bos_19, 
           height=10, aspect=2, scatter_kws={'linewidths':.11,'edgecolor':'k'}, legend=False)

plt.xlabel('Age')
plt.ylabel('Time (min)')
plt.xlim(15, 85)
plt.ylim(115, 475)

plt.legend(labels=['Men','Women'], frameon=False, loc='upper right', bbox_to_anchor=(1, .9))
sns.despine(offset=10, trim=False)

plt.savefig('3__2019_actual_finish_times_by_age_both.png');
```

```python
#Age-Graded Finish Time by Age (Men & Women, Boston 2019)
sns.lmplot(x="Age", y="2020_AG_Result_minutes", hue="Gender", data=bos_19, 
           height=10, aspect=2, scatter_kws={'linewidths':.11,'edgecolor':'k'}, legend=False)

plt.xlabel('Age')
plt.ylabel('Time (min)')
plt.xlim(15, 85)
plt.ylim(115, 475)

plt.legend(labels=['Men','Women'], frameon=False, loc='upper right', bbox_to_anchor=(1, .9))
sns.despine(offset=10, trim=False)

plt.savefig('4__2019_actual_finish_times_by_age_both.png');
```

#### Part III. Boston Qualifying Times

```python
bqs=pd.read_excel('../Too_Large/marathon_standards_change_analysis.xlsx',
                 sheet_name='BQs_over_time_AG20', skiprows=3)
```

```python
#Boston Marathon Qualifying Times (Men & Women, Boston 2003-2020)
plt.figure(figsize=(20,10))

#Women 2003-2012
sns.regplot(x='Age', y='Qualifying_Time_Women_03_12_min',
            color='lightsalmon', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")
#Women 2013-2019
sns.regplot(x='Age', y='Qualifying_Time_Women_13_19_min',
            color='tomato', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")
#Women 2020
sns.regplot(x='Age', y='Qualifying_Time_Women_2020_min',
            color='red', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")



#Men 2003-2012
sns.regplot(x='Age', y='QT_Men_03_12_min',
            color='deepskyblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

#Men 2013-2019
sns.regplot(x='Age', y='QT_Men_13_19_min',
            color='dodgerblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

#Men 2020
sns.regplot(x='Age', y='QT_Men_2020_min',
            color='darkblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

plt.xlabel('Age')
plt.ylabel('Time (min)')

plt.ylim(125, 450)
plt.xlim(15, 85)

plt.legend(labels=['Women (2003-12)','Women (2013-19)', 'Women (2020)', 'Men (2003-12)', 'Men (2013-19)', 'Men (2020)'], frameon=False, loc=2)
sns.despine(offset=10, trim=False)

plt.savefig('5__BQ_times_2003_to_2020.png');
```

```python
#Boston Marathon Qualifying Times (Men & Women, Boston 2020)
plt.figure(figsize=(20,10))

sns.regplot(x='Age', y='Qualifying_Time_Women_2020_min',
            color='red', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

sns.regplot(x='Age', y='QT_Men_2020_min',
            color='darkblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

#plt.title('Actual Finish Time by Age', fontsize=30)
plt.xlabel('Age')
plt.ylabel('Time (min)')

plt.ylim(125, 450)
plt.xlim(15, 85)

plt.legend(labels=['Women','Men'], frameon=False, loc='upper right', bbox_to_anchor=(1, .95))
sns.despine(offset=10, trim=False)

plt.savefig('6__BQ_times_2020.png');
```

```python
#Age-Graded Boston Marathon Qualifying Times (Men & Women, Boston 2020)
plt.figure(figsize=(20,10))

sns.regplot(x='Age', y='AG_Time_Women_2020_min',
            color='red', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

sns.regplot(x='Age', y='AG_Time_Men_2021',
            color='darkblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

plt.xlabel('Age')
plt.ylabel('Time (min)')

plt.ylim(125, 450)
plt.xlim(15, 85)

plt.legend(labels=['Women','Men'], frameon=False, loc='upper right', bbox_to_anchor=(1, .95))

sns.despine(offset=10, trim=False)

plt.savefig('7__AG_BQ_times_2020.png');
```

#### Part IV.  Boston Qualifying Times and 2019 Finish Times

```python
#MEN AG BQs AND AG Boston 2019 Finish Times

plt.figure(figsize=(20,10))

#AG Boston 2019 Finish Times
sns.regplot(x='Age', y='2020_AG_Result_minutes',
            color='tab:blue', data=bos_19[bos_19['Gender'].isin(['M'])], scatter_kws={'s':20}, ci=None) 

#AG BQ times
sns.regplot(x='Age', y='AG_Time_Men_2021',
            color='darkblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

plt.xlabel('Age')
plt.ylabel('Time (min)')

plt.ylim(125, 450)
plt.xlim(15, 85)

plt.legend(labels=['Finish Times', 'Qualifying Times'], frameon=False, loc='upper right', bbox_to_anchor=(1.1, .9))
sns.despine(offset=10, trim=False)

plt.savefig('8__AG_BQs_and_AG_Bos19_Finish_Times_MEN.png');
```

```python
#WOMEN AG BQs AND AG Boston 2019 Finish Times

plt.figure(figsize=(20,10))

sns.regplot(x='Age', y='2020_AG_Result_minutes',
            color='tab:orange', data=bos_19[bos_19['Gender'].isin(['F'])], scatter_kws={'s':20}, ci=None) 

sns.regplot(x='Age', y='AG_Time_Women_2020_min',
            color='red', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

plt.xlabel('Age')
plt.ylabel('Time (min)')

plt.ylim(125, 450)
plt.xlim(15, 85)

plt.legend(labels=['Finish Times', 'Qualifying Times'], frameon=False, loc='upper right', bbox_to_anchor=(1.1, .9))
sns.despine(offset=10, trim=False)

plt.savefig('9__AG_BQs_and_AG_Bos19_Finish_Times_WOMEN.png');
```

#### Part V.  Proposed Boston Marathon Qualifying Times Using Age Grading

```python
bos_19['2020_AG_%'] =bos_19['2020_AG_%']*100
bos_19['2020_AG_%'].head()
```

```python
bqs['AG%_Men_2020']=bqs['AG%_Men_2020']*100
bqs['AG%_Men_2020'].head()
```

```python
bqs['AG%_Women_2020'] =bqs['AG%_Women_2020']*100
bqs['AG%_Women_2020'].head()
```

```python
bqs['Percentage_diff_from_existing_M']=bqs['Percentage_diff_from_existing_M']*100
bqs['Percentage_diff_from_existing_M'].head()
```

```python
bqs['Percentage_diff_from_existing_W']=bqs['Percentage_diff_from_existing_W']*100
bqs['Percentage_diff_from_existing_W'].head()
```

```python
bqs['Proposed_AG%_men']=bqs['Proposed_AG%_men']*100
bqs['Proposed_AG%_men'].head()
```

```python
bqs['Proposed_AG%_women']=bqs['Proposed_AG%_women']*100
bqs['Proposed_AG%_women'].head()
```

```python
#Proposed Boston Marathon Qualifying Times (Men & Women)
plt.figure(figsize=(20,10))

sns.regplot(x='Age', y='Proposed_QT_W_Agegrouped_roundedup_min',
            color='red', data=bqs, scatter_kws={'s':100}, ci=None, marker="D")

sns.regplot(x='Age', y='Proposed_QT_M_Agegrouped_roundedup_min',
            color='darkblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="D")

plt.xlabel('Age')
plt.ylabel('Time (min)')

plt.ylim(125, 450)
plt.xlim(15, 85)

plt.legend(labels=['Women','Men'], frameon=False, loc=1)
sns.despine(offset=10, trim=False)

#plt.savefig('Proposed_BQ_times.png');
```

```python
#Time Diff (min) b/w proposed and actual BQs MEN
plt.figure(figsize=(20,10))
ax = sns.barplot(x="Age_group", y="Time_diff_from_existing_M", data=bqs, color='tab:blue')

plt.axhline(0, color='black', linewidth=1)
plt.xlabel('Age Group')
plt.ylabel('Time (min)')
sns.despine(offset=10, trim=False)

plt.rc('xtick', labelsize=18)
plt.ylim(-15, 70);

#plt.savefig('MEN_time_diff_min_proposed_actual_BQ.png');
```

```python
#Time Diff (min) b/w proposed and actual BQs Women
plt.figure(figsize=(20,10))
ax = sns.barplot(x="Age_group", y="Time_diff_from_existing_W", data=bqs, color='darkorange')

plt.axhline(0, color='black', linewidth=1)
plt.xlabel('Age Group')
plt.ylabel('Time (min)')
sns.despine(offset=10, trim=False)

plt.rc('xtick', labelsize=18)
plt.ylim(-15, 70);
#plt.savefig('WOMEN_time_diff_min_proposed_actual_BQ.png');
```

```python
#Time Diff (%) b/w proposed and actual BQs Men
plt.figure(figsize=(20,10))
ax = sns.barplot(x="Age_group", y="Percentage_diff_from_existing_M", data=bqs, color='tab:blue')

plt.axhline(0, color='black', linewidth=1)
plt.xlabel('Age Group')
plt.ylabel('Percent')
sns.despine(offset=10, trim=False)

plt.rc('xtick', labelsize=18)
plt.ylim(-10, 25);
#plt.savefig('MEN_time_diff_%_proposed_actual_BQ.png');
```

```python
#Time Diff (%) b/w proposed and actual BQs Women
plt.figure(figsize=(20,10))
ax = sns.barplot(x="Age_group", y="Percentage_diff_from_existing_W", data=bqs, color='darkorange')

plt.axhline(0, color='black', linewidth=1)

plt.xlabel('Age Group')
plt.ylabel('Percent')
sns.despine(offset=10, trim=False)

plt.rc('xtick', labelsize=18)
plt.ylim(-10, 25);
#plt.savefig('WOMEN_time_diff_%_proposed_actual_BQ.png');
```

```python
#Boston Qualifying Times Age-Grade Percentages with AG% performance levels

plt.figure(figsize=(20,10))
ax = sns.lineplot(x="Age", y="AG%_Women_2020",
                     data=bqs, linewidth=4, color='red')
ax = sns.lineplot(x="Age", y="AG%_Men_2020",
                     data=bqs, linewidth=4, color='darkblue')

ax.axhline(60, color='black', linewidth=1)
ax.axhline(70, color='black', linewidth=1)
ax.axhline(80, color='black', linewidth=1)
ax.axhline(90, color='black', linewidth=1)

ax.text(15, 95, "World Class", horizontalalignment='left', size='medium', color='gray')
ax.text(15, 85, "National Class", horizontalalignment='left', size='medium', color='gray')
ax.text(15, 75, "Regional Class", horizontalalignment='left', size='medium', color='gray')
ax.text(15, 66, "Local Class", horizontalalignment='left', size='medium', color='gray')

plt.legend(labels=['Women','Men'], frameon=False, loc='upper right', bbox_to_anchor=(1, .95))
plt.rc('xtick', labelsize=22)   
plt.xlabel('Age')
plt.ylabel('Percent')
plt.ylim(50, 100)
#plt.xlim(15, 85)
sns.despine(offset=10, trim=False)

plt.savefig('10__BQ_times_AG%.png');
```

```python
#PROPOSED Boston Qualifying Times Age-Grade Percentages with AG% performance levels

plt.figure(figsize=(20,10))
ax = sns.lineplot(x="Age", y="Proposed_AG%_women",
                     data=bqs, linewidth=4, color='red')
ax = sns.lineplot(x="Age", y="Proposed_AG%_men",
                     data=bqs, linewidth=4, color='darkblue')

ax.axhline(60, color='black', linewidth=1)
ax.axhline(70, color='black', linewidth=1)
ax.axhline(80, color='black', linewidth=1)
ax.axhline(90, color='black', linewidth=1)

ax.text(15, 95, "World Class", horizontalalignment='left', size='medium', color='gray')
ax.text(15, 85, "National Class", horizontalalignment='left', size='medium', color='gray')
ax.text(15, 75, "Regional Class", horizontalalignment='left', size='medium', color='gray')
ax.text(15, 65, "Local Class", horizontalalignment='left', size='medium', color='gray')

plt.legend(labels=['Women','Men'], frameon=False, loc='upper right', bbox_to_anchor=(1, .95))
plt.rc('xtick', labelsize=22)   
plt.xlabel('Age')
plt.ylabel('Percent')
plt.ylim(50, 100)
#plt.xlim(15, 85)
sns.despine(offset=10, trim=False)

plt.savefig('12__proposed_BQ_times_AG%.png');
```
