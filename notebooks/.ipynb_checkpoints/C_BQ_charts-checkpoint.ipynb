### C. Charts showing Boston Marathon qualifying times

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
%matplotlib inline
```

```python
bqs=pd.read_excel('../Too_Large/marathon_standards_change_analysis.xlsx',
                 sheet_name='BQs_over_time_AG20', skiprows=3)
```

```python
bqs.dtypes
```

```python
# various formatting for charts
#plt.rc('font', size=22)          # controls default text sizes
#plt.rc('axes', titlesize=22)     # fontsize of the axes title
#plt.rc('axes', labelsize=28)    # fontsize of the x and y labels
#plt.rc('xtick', labelsize=22)    # fontsize of the tick labels
#plt.rc('ytick', labelsize=22)    # fontsize of the tick labels
#plt.rc('legend', fontsize=22)    # legend fontsize
#plt.rc('figure', titlesize=44)
```

```python
plt.rc('font', size=22)          
plt.rc('axes', labelsize=28)    
plt.rc('figure', titlesize=44)
plt.rcParams['font.sans-serif'] = "Lucida Sans Unicode"
plt.figure(figsize=(20,10));
```

```python
#Boston Marathon Qualifying Times (Men & Women, Boston 2003-2020)
plt.figure(figsize=(20,10))

#Women 2003-2012
sns.regplot(x='Age', y='Qualifying_Time_Women_03_12_min',
            color='lightsalmon', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")
#Women 2013-2019
sns.regplot(x='Age', y='Qualifying_Time_Women_13_19_min',
            color='tomato', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")
#Women 2020
sns.regplot(x='Age', y='Qualifying_Time_Women_2020_min',
            color='red', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")



#Men 2003-2012
sns.regplot(x='Age', y='QT_Men_03_12_min',
            color='deepskyblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

#Men 2013-2019
sns.regplot(x='Age', y='QT_Men_13_19_min',
            color='dodgerblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

#Men 2020
sns.regplot(x='Age', y='QT_Men_2020_min',
            color='darkblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

plt.xlabel('Age')
plt.ylabel('Time (min)')

plt.ylim(125, 450)
plt.xlim(15, 85)

plt.legend(labels=['Women (2003-12)','Women (2013-19)', 'Women (2020)', 'Men (2003-12)', 'Men (2013-19)', 'Men (2020)'], frameon=False, loc=2)
sns.despine(offset=10, trim=False)
plt.grid(False)

plt.savefig('C1_BQ_times_overtime.png');
```

```python
#Boston Marathon Qualifying Times (Men & Women, Boston 2020)
plt.figure(figsize=(20,10))

sns.regplot(x='Age', y='Qualifying_Time_Women_2020_min',
            color='red', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

sns.regplot(x='Age', y='QT_Men_2020_min',
            color='darkblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

#plt.title('Actual Finish Time by Age', fontsize=30)
plt.xlabel('Age')
plt.ylabel('Time (min)')

plt.ylim(125, 450)
plt.xlim(15, 85)

plt.legend(labels=['Women','Men'], frameon=False, loc='upper right', bbox_to_anchor=(1, .95))
sns.despine(offset=10, trim=False)
plt.grid(False)

plt.savefig('C2_BQ_times_2020.png');
```

```python
#Age-Graded Boston Marathon Qualifying Times (Men & Women, Boston 2020)
plt.figure(figsize=(20,10))

sns.regplot(x='Age', y='AG_Time_Women_2020_min',
            color='red', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

sns.regplot(x='Age', y='AG_Time_Men_2021',
            color='darkblue', data=bqs, scatter_kws={'s':100}, ci=None, marker="X")

plt.xlabel('Age')
plt.ylabel('Time (min)')

plt.ylim(125, 450)
plt.xlim(15, 85)

plt.legend(labels=['Women','Men'], frameon=False, loc='upper right', bbox_to_anchor=(1, .95))

sns.despine(offset=10, trim=False)
plt.grid(False)

plt.savefig('C3_AG_BQ_times_2020.png');
```
