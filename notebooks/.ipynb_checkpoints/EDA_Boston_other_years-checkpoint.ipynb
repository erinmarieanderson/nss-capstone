```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("white") #don't want gridlines
%matplotlib inline
```

```python
plt.rc('font', size=22)          
plt.rc('axes', labelsize=28)    
plt.rc('figure', titlesize=44)
plt.rcParams['font.sans-serif'] = "Lucida Sans Unicode"
```

### 2018

```python
bos_18=pd.read_excel('../data/boston_marathon_results_2018.xlsx',
                 sheet_name='bos_18', usecols=['AgeOnRaceDay', 'Gender', 'OfficialTime_min', '2020_AG_Result_min'])
```

```python
bos_18.head()
```

```python
#Actual Finish Time by Age (Men & Women, Boston 2018)
sns.lmplot(x="AgeOnRaceDay", y="OfficialTime_min", hue="Gender", data=bos_18, 
           height=10, aspect=2, scatter_kws={'linewidths':.11,'edgecolor':'k'}, legend=False)

plt.xlabel('Age')
plt.ylabel('Time (min)')
plt.xlim(15, 85)
plt.ylim(115, 475)

plt.legend(labels=['Men','Women'], frameon=False, loc='upper right', bbox_to_anchor=(1, .9))
sns.despine(offset=10, trim=False)

#plt.savefig('3__2019_actual_finish_times_by_age_both.png');
```

```python
#Age-Graded Finish Times by Age (Men & Women, Boston 2018)
sns.lmplot(x="AgeOnRaceDay", y="2020_AG_Result_min", hue="Gender", data=bos_18, 
           height=10, aspect=2, scatter_kws={'linewidths':.11,'edgecolor':'k'}, legend=False)

plt.xlabel('Age')
plt.ylabel('Time (min)')
plt.xlim(15, 85)
plt.ylim(115, 475)

plt.legend(labels=['Men','Women'], frameon=False, loc='upper right', bbox_to_anchor=(1, .9))
sns.despine(offset=10, trim=False)

#plt.savefig('3__2018_AG_finish_times_by_age_both.png');
```
